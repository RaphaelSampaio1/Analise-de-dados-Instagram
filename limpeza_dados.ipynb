```python
import pandas as pd
import re

# importar base de dados suja
base = pd.read_excel('dados_instagram.xlsx')
display(base.head(15))
```

```python
# LIMPANDO BASE

# MOSTAR ESPA√áOS EM BRANCO DESCRI√á√ÉO
base['Descri√ß√£o'] = base['Descri√ß√£o'].replace(r'\n',' ',regex=True)
# CRIAR COLUNA COM HOR√ÅRIO
base['Horario'] = base['Data Publica√ß√£o'].str.extract(r'(\d{2}:\d{2})')
base['Data Publica√ß√£o'] = base['Data Publica√ß√£o'].str.replace(r'\d{2}:\d{2}', '').str.strip()

# REMOVER DIA E 'DE' DA COLUNA DE DATA
dias = ([r'Segunda','Ter√ßa','Quarta','Quinta','Sexta','S√°bado','Domingo',' ',',','\d{2}:\d{2}'])
for dia in dias:
    base['Data Publica√ß√£o'] = base['Data Publica√ß√£o'].str.replace(dia,'',regex=True)

# COLOCAR / NAS DATAS
base['Data Publica√ß√£o'] = base['Data Publica√ß√£o'].replace('de','/',regex=True)
meses = {
    'Janeiro': '01', 'Fevereiro': '02', 'Mar√ßo': '03', 'Abril': '04', 'Maio': '05', 'Junho': '06',
    'Julho': '07', 'Agosto': '08', 'Setembro': '09', 'Outubro': '10', 'Novembro': '11', 'Dezembro': '12',
    'jan': '01', 'fev': '02', 'mar': '03', 'abr': '04', 'mai': '05', 'jun': '06',
    'jul': '07', 'ago': '08', 'set': '09', 'out': '10', 'nov': '11', 'dez': '12','ho':'','o':'','il':''
}
# TROCAR DESCRI√á√ÉO DE MES POR NUMERO
for mes, num in meses.items():
    base['Data Publica√ß√£o'] = base['Data Publica√ß√£o'].str.replace(mes,num,regex=True)
base['Data Publica√ß√£o'] = base['Data Publica√ß√£o'].str.replace(' ','',regex=True)

# COLOCANDO O ANO
base['Data Publica√ß√£o'] = base['Data Publica√ß√£o'] + '/2024'

# TRANSFORMAR EM DATA
base['Data Publica√ß√£o'] = pd.to_datetime(base['Data Publica√ß√£o'],format='%d/%m/%Y')


# SUBSTITUIR VALOR NULO (--) POR 0 (S√ÉO FOTOS)
base[['Reprodu√ß√µes','Minutos Assistidos']] = base[['Reprodu√ß√µes','Minutos Assistidos']].replace('‚Äë‚Äë',0)
#ARRUMAR QUANTIDADE REPRODU√á√ïES

# NOVA ORDEM DE BASE REMOVENDO INTERA√á√ïES
base = base[['Descri√ß√£o','Data Publica√ß√£o','Horario','Alcances','Impress√µes','Curtidas/Rea√ß√µes','Comentarios','Reprodu√ß√µes','Minutos Assistidos']]

display(base.head(15))
```

```python
# SALVAR BASE LIMPA
salvar = base.to_excel('Instagram_clean_data.xlsx',index=False)
```
